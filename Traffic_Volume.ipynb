```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Function to generate sample traffic data
def generate_traffic_data(start, end, cars, bikes, buses, trucks, autos, duration):
    dates = pd.date_range(start="2024-06-20", periods=1, freq='D')
    times = pd.date_range(start=start, end=end, freq='15T').time
    
    data = []
    for date in dates:
        for time in times:
            count_cars = np.random.randint(cars[0], cars[1])
            count_two_wheeler = np.random.randint(bikes[0], bikes[1])
            count_bus = np.random.randint(buses[0], buses[1])
            count_trucks = np.random.randint(trucks[0], trucks[1])
            count_autos = np.random.randint(autos[0], autos[1])
            vehicle_count = count_cars + count_bus + count_two_wheeler + count_trucks + count_autos
            average_travel_time = np.random.uniform(duration[0], duration[1]) # in minutes
        
            data.append([date, time, count_cars, count_two_wheeler, count_bus, count_trucks, count_autos, vehicle_count, average_travel_time])
            
    return pd.DataFrame(data, columns=["Date", "Start-Time", "Cars", "Bikes", "Buses", "Trucks", "Autos", "Total", "Travel Time (minutes)"])
```

```python
# Generate data for three roads
sion_panvel_data_01 = generate_traffic_data("05:30", "08:00", [50, 75], [50, 75], [10, 15], [0, 10], [20, 30], [50, 60])
sion_panvel_data_02 = generate_traffic_data("08:15", "11:00", [200, 250], [200, 250], [10, 15], [0, 10], [40, 50], [90, 100])
sion_panvel_data_03 = generate_traffic_data("11:15", "16:00", [100, 150], [100, 150], [10, 15], [0, 10], [30, 40], [60, 70])
sion_panvel_data_04 = generate_traffic_data("16:15", "21:00", [200, 250], [200, 250], [10, 15], [0, 10], [40, 50], [90, 100])
sion_panvel_data_05 = generate_traffic_data("21:15", "23:00", [50, 75], [50, 75], [10, 15], [0, 10], [20, 30], [50, 60])
```

```python
sion_panvel_data = pd.concat([sion_panvel_data_01, sion_panvel_data_02, sion_panvel_data_03, sion_panvel_data_04, sion_panvel_data_05], ignore_index = True)
```

```python
bandra_dahisar_data_01 = generate_traffic_data("05:30", "08:00", [50, 75], [50, 75], [10, 15], [0, 10], [20, 30], [50, 60])
bandra_dahisar_data_02 = generate_traffic_data("08:15", "11:00", [200, 250], [200, 250], [10, 15], [0, 10], [40, 50], [90, 100])
bandra_dahisar_data_03 = generate_traffic_data("11:15", "16:00", [100, 150], [100, 150], [10, 15], [0, 10], [30, 40], [60, 70])
bandra_dahisar_data_04 = generate_traffic_data("16:15", "21:00", [200, 250], [200, 250], [10, 15], [0, 10], [40, 50], [90, 100])
bandra_dahisar_data_05 = generate_traffic_data("21:15", "23:00", [50, 75], [50, 75], [10, 15], [0, 10], [20, 30], [50, 60])
```

```python
bandra_dahisar_data = pd.concat([bandra_dahisar_data_01, bandra_dahisar_data_02, bandra_dahisar_data_03, bandra_dahisar_data_04, bandra_dahisar_data_05], ignore_index = True)
```

```python
def bar_plot(road, index, vehicle):
    plt.style.use("ggplot")
    plt.figure(figsize = (8,5))
    plt.grid(True)
    sns.barplot(data = df[df["Road"] == road].iloc[index:index+4], x = "Start-Time", y = vehicle)
    plt.xlabel("Time Interval", fontsize = 10)
    plt.xticks(fontsize = 8)
    plt.ylabel("No. of " + vehicle, fontsize = 10)
    plt.yticks(fontsize = 8)
    plt.title("Time Interval v/s " + vehicle, fontsize = 15, loc = "center")
    plt.show()
```

```python
def pie_plot(road, index):
    cars = df[df["Road"] == road].iloc[index:index+4]["Cars"].sum()
    bikes = df[df["Road"] == road].iloc[index:index+4]["Bikes"].sum()
    buses = df[df["Road"] == road].iloc[index:index+4]["Buses"].sum()
    autos = df[df["Road"] == road].iloc[index:index+4]["Autos"].sum()
    trucks = df[df["Road"] == road].iloc[index:index+4]["Trucks"].sum()
    
    values = [cars, bikes, buses, autos, trucks]
    labels = ["Cars", "Bikes", "Buses", "Autos", "Trucks"]
    
    def make_autopct(values):
        def my_autopct(pct):
            total = sum(values)
            val = int(round(pct*total/100.0))
            return '{p:.2f}%'.format(p=pct,v=val)
        return my_autopct
    
    plt.style.use("ggplot")
    plt.figure(figsize = (7,7))
    plt.grid("both")
    fig = plt.pie(x = values, labels = labels, autopct = make_autopct(values), pctdistance = 0.7, radius = 0.9)
    plt.title(road + " - Percentage of Vehicles in Peak Hour", fontsize = 15, fontstyle = "italic", loc = "center")
    plt.legend(shadow = True)
    plt.show()
```

```python
pie_plot("Sion-Panvel Road", index_01)
pie_plot("Andheri-Dadar Road", index_02)
pie_plot("Bandra-Dahisar Road", index_03)
```
